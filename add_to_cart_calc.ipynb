```python
from sqlalchemy import create_engine, text
import pandas as pd
```

```python
def atc_pct_calc(schema,
                 table_all,
                 table_atc,
                 pg_ts_user = 'postgres',
                 pg_ts_password = 'a5TKQ7SSVGFuQIz5',
                 pg_ts_hostname = 'primeds.protal.biz:30308',
                 pg_ts_db_name = 'postgres'):
    
    pg_engine = create_engine(f'postgresql+psycopg2://{pg_ts_user}:{pg_ts_password}@{pg_ts_hostname}/{pg_ts_db_name}')
    pg_conn = pg_engine.connect()

    query1 = f"SELECT * FROM {schema}.{table_all}"
    df_all = pd.read_sql(text(query1), pg_conn)

    query2 = f"SELECT * FROM {schema}.{table_atc}"
    df_atc = pd.read_sql(text(query2), pg_conn)

    df_diff = df_all.merge(df_atc, on="event_time")

    df_diff["atc_pct"] = (df_diff["user_count_y"] / df_diff["user_count_x"]) * 100
    df_diff = df_diff.drop(["user_count_x", "user_count_y"], axis="columns")

    pg_conn.close()

    return df_diff
```

```python
atc_pct_calc(schema='staging', table_all='daily_users', table_atc='daily_users_add_to_cart')
```
